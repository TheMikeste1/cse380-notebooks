```
from math import ceil, log
from collections import Counter

def show_results(message, code_tuples):
  total_characters = len(message)
  total_unique_characters = len(code_tuples)
  total_bits = 0
  for char, count, code in code_tuples:
    total_bits += count * len(code)
  average_bits_per_character = total_bits / total_characters
  fixed_bits_per_character = ceil(log(total_unique_characters, 2))
  total_fixed_bits = total_characters * fixed_bits_per_character
  compression_ratio = (total_fixed_bits - total_bits) / total_fixed_bits
  print(f'          Total Characters: {total_characters}')
  print(f'   Total Unique Characters: {total_unique_characters}')
  print(f'                Total Bits: {total_bits}')
  print(f'Average Bits per Character: {average_bits_per_character:.2f}')
  print(f'  Fixed Bits per Character: {fixed_bits_per_character}')
  print(f'          Total Fixed Bits: {total_fixed_bits}')
  print(f'         Compression Ratio: {compression_ratio:.3f}')

message1 = 'thebookofmormon'
counter1 = Counter(message1)

print(message1, '-->', counter1)

message2 = 'therestoration'

counter2 = Counter(message2)

print(message2, '-->', counter2)
```

```
from sympy import primerange

list_of_gaps = []
prev = 3
gap = 0
for prime in list(primerange(4, 101)):
    gap = prime - prev
    #print(gap)
    prev = prime
    list_of_gaps.append(gap)

print(list_of_gaps)
```

### Response

We'll first make a few functions and test them.

```
from collections import defaultdict 

occurrences = defaultdict(int)
for gap in list_of_gaps:
  occurrences[str(gap)] += 1
occurrences
```

```
def get_huffman_codes(occurrences):
    occurrences = list(occurrences)
    codes = defaultdict(str)
    while len(occurrences) > 1:
        # Sort the occurrences so the least common items are first
        occurrences = sorted(occurrences, key=lambda x: (x[1], x[0]))

        # Pop 2 at a time to create a new node
        node_a = occurrences.pop(0)
        node_b = occurrences.pop(0)

        for gap_size in node_a[0].split('-'):
            codes[str(gap_size)] += '0'
        for gap_size in node_b[0].split('-'):
            codes[str(gap_size)] += '1'

        new_node = (node_a[0] + '-' + node_b[0], node_a[1] + node_b[1])
        occurrences.append(new_node)

    return codes
```

```
def create_tuples(codes, occurrences):
  tuples = []
  for gap, code in codes.items():
    t = (gap, occurrences[gap], code)
    tuples.append(t)
  return tuples
```

```
codes = get_huffman_codes(occurrences.items())
gap_tuples = create_tuples(codes, occurrences)
show_results(list_of_gaps, gap_tuples)
```

Now that we have those functions working, we can try different numbers of primes until we have a compression ratio better than 24%.

```
def calculate_compression_ratio(message, code_tuples):
  total_characters = len(message)
  total_unique_characters = len(code_tuples)
  total_bits = 0
  for char, count, code in code_tuples:
    total_bits += count * len(code)
  
  fixed_bits_per_character = ceil(log(total_unique_characters, 2))
  total_fixed_bits = total_characters * fixed_bits_per_character
  compression_ratio = (total_fixed_bits - total_bits) / total_fixed_bits

  return compression_ratio
```

```
import sympy

cr = 0
primes = list(primerange(4, 101))
while cr <= 0.24:
  # Get the next prime
  # We have to add three because we're skipping
  # the first 2 primes, and we want the next prime.
  primes.append(sympy.prime(len(primes) + 3))

  # Get Gaps
  list_of_gaps = []
  prev = 3
  gap = 0
  for prime in primes:
    gap = prime - prev
    prev = prime
    list_of_gaps.append(gap)
  
  # Generate occurrences
  occurrences = defaultdict(int)
  for gap in list_of_gaps:
    occurrences[str(gap)] += 1

  codes = get_huffman_codes(occurrences.items())
  gap_tuples = create_tuples(codes, occurrences)
  cr = calculate_compression_ratio(list_of_gaps, gap_tuples)
  print(cr)
  
print(len(primes), "primes were required.")
print(primes)
```
