### In terms of the Binomial Theorem's $(1 + x)^n$ reduced mod 2

In the following, ```rm2``` stands for 'reduced mod 2'.

1. Expand and reduce: $$(1 + x)^2 = 1 + 2x + x^2,\ \mbox{rm2} \Rightarrow 1 + 0x + x^2 = 1 + x^2$$
2. Try with an odd power: $$(1 + x)^3 = (1 + 2x + x^2) + (x + 2x^2 + x^3) = 1 + 3x + 3x^2 + x^3,\ \mbox{rm2} \Rightarrow 1 + 1x + 1x^2 + x^3$$
3. Better stick with even: $$(1 + x)^4 = 1 + 4x + 6x^2 + 4x^3 + x^4,\ \mbox{rm2} \Rightarrow 1 + x^4$$
4. Or decomposed into power of 2 powers: $$(1 + x)^{11} = (1 + x)^{8 + 2 + 1} = (1 + x)^8 (1 + x)^2 (1 + x)^1,\ \mbox{rm2} \Rightarrow (1 + x^8)(1 + x^2)(1 + x^1)$$

The number of evens is $2^x$, where $x$ is the number of zeroes to the right of the most major 1.

The most-even rows:

 | n  | n in binary | no. of 0's in n's binary rep. | $2^{\mbox{no. of 0's}}$  |
 |----|-------------|-------------------------------|--------------------------|
 | 2  |          10 |                             1 | $2^1 = 2$                |
 | 4  |         100 |                             2 | $2^2 = 4$                |
 | 8  |        1000 |                             3 | $2^3 = 8$                |
 | 16 |       10000 |                             4 | $2^4 = 16$               |

The total number of odds quickly becomes dwarfed by the total number of evens, meaning the chance of choosing an even number approaches 100%.

Another way to calculate it:
$$
P(Even) = 1 - \lim_{n \rightarrow \infty} P_n(Odd)
$$
where $n$ equals the number of rows.

Note that the total number of odds is equal to $\lfloor 3^{\log_2{n}} \rfloor$, where $n$ is the row number, when $n$ is a power of 2 (e.g., 8, 16, 32).

````
from math import log2, floor
from fractions import Fraction
````

````
floor(3 ** log2(7))
````

The total number of values is equal to the triangle number for row $n$, which is calculated by $\frac{n(n+1)}{2}$.

$$
P(Even) = 1 - \lim_{n \rightarrow \infty} P_n(Odd) =
1 - \lim_{n \rightarrow \infty} \frac{3^{\log_2{n}}}{\frac{n(n + 1)}{2}} =
1 - 2\lim_{n \rightarrow \infty} \frac{3^{\log_2{n}}}{n^2 + n}
$$

````
def odds_over_total(n):
  odds = floor(3 ** log2(n))
  total = floor((n * (n + 1)) / 2)
  return Fraction(odds, total)
````

````
frac = odds_over_total(2 ** 10)
print(frac, "=", float(frac))
frac = odds_over_total(2 ** 100)
print(frac, "=", float(frac))
````

As $n$ approaches $\infty$, the limit approaches $0$. Therefore, $P(Even) = 1 - 0 = 1 = 100\%$.
